# Geração de espectrogramas para detecção de palmas

Esse notebook será usado para gerar espectrogramas dos áudios usados como base de dados com o objetivo de identificar ruídos (palmas) para elimina-los da base de dados.

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import os

%matplotlib inline
```

```python
audio_path = '../Data/Audio/Segmented/'
spec_path = '../Data/Spectrogram/Segmented/'

files = os.listdir(audio_path)

plt.figure(figsize=(12,8))

for f in files:
    
    plt.clf()
    
    print("Processing File", f)
    audio_file_path = os.path.join(audio_path, f)
    
    # Loading with Librosa
    
    y, sr = librosa.load(audio_file_path)
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
    
    # Generating spectrogram
    
    librosa.display.specshow(D, y_axis='linear')
    
    # Setting up spectrogram configs

    plt.colorbar(format='%+2.0f dB')
    plt.title(f)
    
    # Saving spectrogram
    
    spec_filename = f[:-4] + '.png'
    spectrogram_file_path = os.path.join(spec_path, spec_filename)
    plt.savefig(spectrogram_file_path) 
```
