```python
import librosa
import librosa.display
import IPython
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Loading data to be transformed using STFT

```python
y, sr = librosa.load(librosa.util.example_audio_file(), duration=5, offset=0)
S = np.abs(librosa.stft(y, center=False))
y_inv = librosa.griffinlim(S)
y_istft = librosa.istft(S)
```

```python
librosa.display.specshow(librosa.amplitude_to_db(S, ref=np.max), sr=sr, y_axis='log',  x_axis='time')
plt.title("Power Spectrogram")
plt.colorbar(format="%+2.0f dB")
plt.tight_layout()
```

### Original Audio Source

```python
IPython.display.Audio(data=y, rate=sr)
```

### Griffin Lim Reconstruction

```python
IPython.display.Audio(data=y_inv, rate=sr)
```

### iSTFT Reconstruction

```python
IPython.display.Audio(data=y_istft, rate=sr)
```

```python
plt.figure(figsize=(10, 1))
librosa.display.waveplot(y, sr=sr)
plt.figure(figsize=(10, 1))
librosa.display.waveplot(y_inv, sr=sr)
plt.figure(figsize=(10, 1))
librosa.display.waveplot(y_istft, sr=sr)
```
